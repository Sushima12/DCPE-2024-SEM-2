```python
import matplotlib.pyplot as plt
import pickle

# Load the data from the updated .pkl file
with open("training_history.pkl", "rb") as file:
    training_data = pickle.load(file)

# Extract data
epochs = list(range(1, len(training_data["accuracy"]) + 1))
accuracy = training_data["accuracy"]
val_accuracy = training_data["val_accuracy"]
loss = training_data["loss"]
val_loss = training_data["val_loss"]

# Adjusted validation accuracy (Optional, same as val_accuracy here)
val_accuracy_adjusted = val_accuracy

# Plotting the graphs
fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharex=True)
fig.patch.set_facecolor("#e6f7ff")

# Subplot 1: Loss
axes[0].plot(epochs, loss, label='Loss', linewidth=2)
axes[0].plot(epochs, val_loss, label='Validation Loss', linewidth=2)
axes[0].set_title('Loss (30 Epochs)', fontsize=14)
axes[0].set_xlabel('Epoch', fontsize=12)
axes[0].set_ylabel('Loss', fontsize=12)
axes[0].legend(fontsize=12)

# Subplot 2: Accuracy
axes[1].plot(epochs, accuracy, label='Accuracy', linewidth=2)
axes[1].plot(epochs, val_accuracy_adjusted, label='Validation Accuracy (Adjusted)', linewidth=2)
axes[1].set_title('Accuracy (30 Epochs)', fontsize=14)
axes[1].set_xlabel('Epoch', fontsize=12)
axes[1].set_ylabel('Accuracy', fontsize=12)
axes[1].legend(fontsize=12)

# Set background and layout
for ax in axes:
    ax.set_facecolor("#e6f7ff")
    ax.tick_params(axis='both', labelsize=10)

plt.tight_layout()
plt.show()
```
